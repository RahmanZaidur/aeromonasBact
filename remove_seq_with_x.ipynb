## Code to remove the sequences which had "X" in them. X means unknown amino acid. After removing, uniprot_df.csv file was updated.

```python
import pandas as pd
```

```python
uniprot_df = pd.read_csv("uniprot_df.csv")
uniprot_df
```

```python
#Find the rows that contains "X" in the sequences

uniprot_df[uniprot_df['sequence'].str.contains('X')]
```

```python
#Remove the rows and update the uniprot_df

uniprot_df = uniprot_df[~uniprot_df['sequence'].str.contains('X', case=False)]
uniprot_df
```

```python
#update the uniprot_df csv file

uniprot_df.to_csv("/Users/zaidur/Documents/Sequence_Project/aeromonasBact/uniprot_df.csv", index=False)
```
