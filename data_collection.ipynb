```python
from Bio.Blast import NCBIWWW, NCBIXML
from Bio import SeqIO

# Load the query sequence from a FASTA file
query_sequence = SeqIO.read("/Users/zaidur/Documents/Sequence Project/aeromonasBact/combined_singleton copy.faa", format="fasta")

# Run a BLAST search using the query sequence against a database
result_handle = NCBIWWW.qblast("blastp", "nr", query_sequence.seq)

# Parse the BLAST results
blast_records = NCBIXML.parse(result_handle)
blast_record = next(blast_records)

# Print the homologous sequences
for alignment in blast_record.alignments:
    for hsp in alignment.hsps:
        print(f"Sequence ID: {alignment.title}")
        print(f"Sequence Length: {alignment.length}")
        print(f"Score: {hsp.score}")
        print("Query sequence:", hsp.query)
        print("Match sequence:", hsp.match)
        print("Alignment start:", hsp.query_start)
        print("Alignment end:", hsp.query_end)
```

```python
from Bio.Blast import NCBIWWW, NCBIXML
from Bio import SeqIO

# Load the protein sequences from a FASTA file
protein_sequences = list(SeqIO.parse("/Users/zaidur/Documents/Sequence Project/aeromonasBact/sample_fast.faa", "fasta"))

# Run a BLAST search for each protein sequence
for query_sequence in protein_sequences:
    # Run a BLAST search using the query sequence against the nr protein database
    result_handle = NCBIWWW.qblast("blastp", "nr", query_sequence.seq)

    # Parse the BLAST results
    blast_records = NCBIXML.parse(result_handle)
    blast_record = next(blast_records)

    # Print the homologous sequences
    for alignment in blast_record.alignments:
        for hsp in alignment.hsps:
            print(f"Query sequence ID: {query_sequence.id}")
            print(f"Sequence ID: {alignment.title}")
            print(f"Sequence Length: {alignment.length}")
            print(f"Score: {hsp.score}")
            print("Query sequence:", hsp.query)
            print("Match sequence:", hsp.match)
            print("Alignment start:", hsp.query_start)
            print("Alignment end:", hsp.query_end)
            print()
```
