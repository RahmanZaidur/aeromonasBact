### Extract the blast features

**Combine the blast hit result log files from cluster and create combine dataframe of 8941896 rows.
Filter the dataframe for hits that have query converage of 100% and identity percentage of 80% or more.
combined the results for each unique sequence ID.
count_blast - total count of blast hits for the sequence
AvgIdBlast - Average of the Identity Percentage of those hits**

```python
#Loading the log files from the blast search result output

import pandas as pd

blast_1 = pd.read_csv("uniprot_blast_results/uniprot_blast_query_1.log", sep='\t', header=None)
blast_1.columns = ["id", "subject_id", "identity_percentage", "alignment_length", "mismatches", "gap_opens", "query_start", "query_end", "subject_start", "subject_end", "expect_value", "bit_score", "query_coverage"]
blast_1
```

```python
#Combining the all of the dataframe into one

# Create a list of your dataframes
dataframes = [blast_1, blast_2, blast_3, blast_4, blast_5, blast_6]

# Use pandas concat() function to combine them
blast_df = pd.concat(dataframes, ignore_index=True)
blast_df
```

```python
#Cobmining the results to create blast features.

grouped = filtered_blast_df.groupby('id')

summary_blast_df = pd.DataFrame()
summary_blast_df['id'] = grouped['id'].first() # This will get the unique ids
summary_blast_df['count_blast'] = grouped.size() # This will count the number of rows for each id
summary_blast_df['AvgIdBlast'] = grouped['identity_percentage'].mean() # This will calculate the mean of 'identity_percentage' for each id

summary_blast_df = summary_blast_df.reset_index(drop=True) # Reset index
summary_blast_df
```

```python
#Remove the sequences to match up with the latest uniprot_df.
#These are removed because they had unknow amino acids as X

# Define the list of ids to remove
ids_to_remove = ['A0A1Q8F8M2', 'A0A428VHZ9', 'P81903', 'Q03321']

# Remove the rows with specified ids
filtered_summary_df = summary_blast_df[~summary_blast_df['id'].isin(ids_to_remove)]

# Print the filtered DataFrame
filtered_summary_df
```
