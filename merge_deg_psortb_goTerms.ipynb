### Combine DEG, PSORTb and GO terms. One-hot encoding the categorical columns

**Combine the DEG, PSORTb and GO terms files. Only keep required columns.Replace the NaN values with 0.
Creat binary columns from the categorical columns**

```python
import pandas as pd
import numpy as np
import ast
```

```python
deg_df = pd.read_csv("/Users/zaidur/Documents/Sequence_Project/aeromonasBact/uniprot_DEG_df.csv")
deg_df
```

```python
psort_df = pd.read_csv("/Users/zaidur/Documents/Sequence_Project/aeromonasBact/psort_extract_df.csv")
psort_df
```

```python
go_df = pd.read_csv("/Users/zaidur/Documents/Sequence_Project/aeromonasBact/go_ancestor_data.csv")
go_df
```

```python
#Reading from csv converts the lists into strings. Make them list again

go_df['go_terms'] = go_df['go_terms'].apply(ast.literal_eval)
go_df['go_parents'] = go_df['go_parents'].apply(ast.literal_eval)
```

```python
# First, let's rename the column 'SeqID' to 'id' in psort_df for consistency
psort_df = psort_df.rename(columns={'SeqID': 'id'})

# Now we can merge the three dataframes
merged_df = deg_df.merge(psort_df, on='id', how='inner').merge(go_df, on='id', how='inner')
```

```python
columns_to_drop = ['sequence_x', 'CMSVM', 'CytoSVM', 'ECSVM', 'ModHMM', 'Motif', 'OMPMotif', 'OMSVM', 
                   'PPSVM', 'Profile', 'SCL-BLAST', 'SCL-BLASTe', 'Signal', 'sequence_y', 'go_terms']

merged_df.drop(columns=columns_to_drop, inplace=True)
```

```python
merged_df
```

```python
merged_df.isna().sum()
```

```python
#Replacing the NaN values with 0

import pandas as pd
import numpy as np

# Replace NaN values with zero in specific columns
merged_df['DEGBacteria_MaxID'].fillna(0, inplace=True)
merged_df['DEGArchea_MaxID'].fillna(0, inplace=True)
merged_df['DEGEukar_MaxID'].fillna(0, inplace=True)

merged_df
```

```python
#one-hot encoding of the categorical column "Final Prediction"

import pandas as pd

# Apply one-hot encoding to "Final Prediction" column
one_hot_encoded = pd.get_dummies(merged_df['Final Prediction'], prefix='prediction_')

# Get the index of the "go_parents" column
go_parents_index = merged_df.columns.get_loc('go_parents')

# Insert the one-hot encoded columns before the "go_parents" column
merged_df = pd.concat([merged_df.iloc[:, :go_parents_index], one_hot_encoded, merged_df.iloc[:, go_parents_index:]], axis=1)

# Drop the original "Final Prediction" column
merged_df.drop('Final Prediction', axis=1, inplace=True)

merged_df
```

```python
#One-hot Encoding of the target Column

import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Assuming merged_df is your DataFrame

# Initialize the MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Fit and transform the 'go_parents' column
transformed_labels = mlb.fit_transform(merged_df['go_parents'])

# Get the label classes from the MultiLabelBinarizer
classes = mlb.classes_

# Create a new DataFrame from the transformed labels and the classes
labels_df = pd.DataFrame(transformed_labels, columns=classes)

# Concatenate the original DataFrame with the new labels DataFrame
final_df = pd.concat([merged_df, labels_df], axis=1)

# Drop the original 'go_parents' column
final_df = final_df.drop(columns=['go_parents'])

final_df
```

```python
"""
Remove the sequences to match up with uniprot_df
'Q9PRY2', 'P81903', 'Q03321', 'A0A428VHZ9', 'A0A1Q8F8M2' are removed because they had X in the sequence
making difficult to get the physiological features
'P86454' was removed because there was no blast hit result for this sequence.
"""


# Define the list of ids to remove
ids_to_remove = ['A0A1Q8F8M2', 'A0A428VHZ9', 'P81903', 'P86454', 'Q03321', 'Q9PRY2']

# Remove the rows with specified ids
final_df = final_df[~final_df['id'].isin(ids_to_remove)]

# Print the filtered DataFrame
final_df
```

```python
final_df.to_csv("DEG_Psort_features.csv", index=False)
```
